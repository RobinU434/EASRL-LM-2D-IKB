```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
target = np.array([-0.5, 0.5])
rad = np.linspace(0, 1, 100)
azm = np.linspace(0, 2 * np.pi, 100)
r, th = np.meshgrid(rad, azm)

cart_coords = np.stack([np.cos(th), np.sin(th)]) * r
cart_coords = cart_coords.T

z = np.linalg.norm(target - cart_coords, axis=2).T / 2

fig = plt.figure()
ax = fig.add_subplot(projection="polar")

imaxes = ax.pcolormesh(th, r, z)

# ax.plot(azm, r, color='k', ls='none') 
fig.colorbar(imaxes)
ax.scatter(np.arctan2(*target[::-1]), np.linalg.norm(target), marker="*", color="r", label="target")
ax.grid()
# axlt.legend(loc="upper left")
ax.set_title("Normalized Reward Function")

fig.savefig("../thesis/figures/methodology/RewardFunction.png")
```
