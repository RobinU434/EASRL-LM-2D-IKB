```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import ndarray
```

```python


def sigmoid(x: ndarray) -> ndarray:
    return 1.0 / (1.0 + np.exp(-x))

def relu(x: ndarray) -> ndarray:
    y = np.zeros_like(x)
    y[x > 0] = x[x > 0]
    return y
```

```python
x = np.linspace(-5, 5, 1000)
```

```python
y = sigmoid(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.set_ylim([-1.1, 1.1])
ax.set_xlim([-5, 5])
ax.set_title("Sigmoid")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
fig.savefig("../thesis/figures/background/sigmoid.png")
```

```python
y = np.tanh(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.set_ylim([-1.1, 1.1])
ax.set_xlim([-5, 5])
ax.set_title("Tanh")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
fig.savefig("../thesis/figures/background/tanh.png")
```

```python
x = np.linspace(-1, 1, 1000)
y = relu(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.set_ylim([-1.1, 1.1])
ax.set_xlim([-1, 1])
ax.set_title("ReLU")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
fig.savefig("../thesis/figures/background/ReLU.png")
```
