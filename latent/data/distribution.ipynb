```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
annotations_file = "../../datasets/5/train/actions_random.csv"
df = pd.read_csv(annotations_file)
print(df.head())

array = np.array(df)[:, 1:].flatten()

mean = array.mean()
std = array.std()

fig = plt.figure()
ax = fig.add_subplot()

num = ax.hist((array) / std, bins=200)
plt.show()

print(mean, std)

```

```python
annotations_file = "../../datasets/10/targets_1674745082.csv"
df = pd.read_csv(annotations_file)
print(df.head())

array = np.array(df)[:, 1:]

x = array[:, 0]
y = array[:, 1]

mean = array.mean()
std = array.std()

fig, axs = plt.subplots(1, 2)
num = axs[0].hist(x, bins=200)
num = axs[1].hist(y, bins=200)

fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(x, y, 1)

plt.show()

print(mean, std)
```
